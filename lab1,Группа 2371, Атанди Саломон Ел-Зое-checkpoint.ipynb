```python
# Example setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Dataset Description

- **Subject Area**: Housing Prices
- **Data Source**: [Kaggle Boston Housing Dataset](https://www.kaggle.com/c/boston-housing)
- **Nature of the Data**: Real data collected from various housing transactions in Boston.

### Attributes
- **CRIM**: Per capita crime rate by town (numeric)
- **ZN**: Proportion of residential land zoned for lots over 25,000 sq. ft. (numeric)
- **INDUS**: Proportion of non-retail business acres per town (numeric)
- **NOX**: Nitric oxides concentration (parts per 10 million) (numeric)
- **RM**: Average number of rooms per dwelling (numeric)
- **AGE**: Proportion of owner-occupied units built prior to 1940 (numeric)
- **DIS**: Weighted distances to five Boston employment centers (numeric)
- **RAD**: Index of accessibility to radial highways (numeric)
- **TAX**: Full-value property tax rate per $10,000 (numeric)
- **PTRATIO**: Pupil-teacher ratio by town (numeric)
- **B**: \(1000(Bk - 0.63)^2\) where \(Bk\) is the proportion of Black residents by town (numeric)
- **LSTAT**: Percentage of lower status of the population (numeric)
- **MEDV**: Median value of owner-occupied homes in $1000s (numeric, target variable)
```

```python
# Load dataset
data = pd.read_csv(r"C:\Users\atand\Downloads\BostonHousing.csv")

# Descriptive Statistics
desc_stats = data.describe()

# Plot histograms
data.hist(bins=20, figsize=(20, 15))
plt.show()

# Missing values
missing_values = data.isnull().sum()

# Handling Missing Values (example: filling with mean)
data.fillna(data.mean(), inplace=True)
```

```python
# Correlation Matrix  
corr_matrix = data.corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Calculate correlation matrix
corr_matrix = data.corr()

# Find highly correlated attributes
high_corr = corr_matrix[(corr_matrix >= 0.7) | (corr_matrix <= -0.7)]
print("Highly correlated attributes (|correlation| > 0.7):")
print(high_corr)
```

```python
low_corr = corr_matrix[(corr_matrix < 0.3) & (corr_matrix > -0.3)]
print("Attributes with low or no correlation (|correlation| < 0.3):")
print(low_corr)
```
